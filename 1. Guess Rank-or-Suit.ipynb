```python
from random import choice
import matplotlib.pyplot as plt
```

```python
SUITS = ["♦", "♣", "♥", "♠"]
RANKS = ["A", "K", "Q", "J", "10", "9", "8", "7", "6", "5", "4", "3", "2"]
DECK = [f"{suit}{rank}" for suit in SUITS for rank in RANKS]
```

```python
def get_random_card():
    return choice(DECK)

def guess_rank():
    return choice(RANKS)

def guess_suit():
    return choice(SUITS)

def guess_rank_or_suit_of_card():
    card = get_random_card()
    return True if guess_suit() in card or guess_rank() in card else False

def mean(l):
    return sum(l) / len(l)
```

```python
def monte_carlo_simulation(n):
    l = []
    for i in range(n):
        l.append(guess_rank_or_suit_of_card())
        
    return mean(l)
```

```python
%%time
SIMULATIONS = list(range(1, 500))
RESULTS = []

for simulation in SIMULATIONS:
    RESULTS.append(monte_carlo_simulation(simulation))
```

##### LET:
###### P(s) - probability of guessing SUIT of CARD
###### P(r) - probability of guessing RANK of CARD
###### P(r $\cap$ s) probability of guessing SUIT AND RANK of CARD
###### P(r $\cup$ s) probability of guessing SUIT OR RANK of CARD

$$P(s) = \frac{1}{4}$$
$$P(r) = \frac{1}{13}$$
$$P(r \cap s) = \frac{1}{52}$$
___
$$P(r \cup s) = P(s) + P(r) - P(r \cap s)$$
$$P(r \cup s) = \frac{1}{4} + \frac{1}{13} - \frac{1}{52} = \frac{16}{52} = \frac{4}{13} \approx 0.308 \approx (30.8 \% )$$
___
Probability of guessing rank or suit of card is $\approx$ 31%

```python
P_r_or_s = 4/13

plt.figure(figsize=(15, 8))
plt.title("Probability of Guessing Suit or Rank of CARD w/ Monte-Carlo Simulation")
plt.ylabel("RATIO of guessing rank or suit")
plt.xlabel("# of Monte-Carlo Simulation Tries")
plt.plot(SIMULATIONS, RESULTS, color='palegreen')
plt.axhline(P_r_or_s, color='#ADD8E6', linestyle = '--', linewidth=7, label=f'P(suit ∪ rank) ≈ {round(P_r_or_s, 6)}')
plt.axhline(mean(RESULTS), color='grey', linestyle = ':', linewidth=2, label=f'Monte-Carlo ≈ {round(mean(RESULTS), 6)}')
plt.legend()
plt.show()
```
